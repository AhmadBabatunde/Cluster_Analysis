## Import the relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
#from sklearn.clusters import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
data = pd.read_csv('Countries-exercise.csv')
```

```python
data.head()
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(data['Longitude'],data['Latitude'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
data_w = data.copy()
data_w = data_w.iloc[:,1:]
data_w.head()
```

## Clustering

Assume there are only two clusters. 

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(7)
x = data_w
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

### Clustering Resutls

```python
data_r = data.copy()
data_r['Cluster By Position'] = identified_clusters
data_r
```

```python
plt.scatter(data_r['Longitude'],data_r['Latitude'],c=data_r['result'],cmap='rainbow')
#plt.ylim(-180,180)
#plt.xlim(-90,90)
plt.show()
```

## Another Example Of Clutstering But This Time With Categorical Data

```python
data2 = pd.read_csv('categorical.csv')
data2.continent.unique()
```

```python
data2['continent'] = data2['continent'].map({'Africa':1,'Asia':2,'Europe':3,'North America':4,'Oceania':5,'Antarctica':6,
                                            'Seven seas (open ocean)':7,'South America':8})
```

```python
p = data2[['continent']]
p.head()
```

```python
kmeans2 = KMeans(n_clusters=7, init='k-means++', max_iter=300)
```

```python
kmeans2.fit(p)
```

```python
result2 = kmeans2.fit_predict(p)
```

```python
data2['Cluster By Continent'] = result2
data2.head(10)
```

```python
plt.scatter(data2['Longitude'],data2['Latitude'],c=data2['Cluster By Continent'],cmap='rainbow')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```
